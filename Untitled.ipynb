```python
rating_matrix_2 = rating_matrix.as_matrix()
user_ratings_mean = np.mean(rating_matrix_2, axis = 1)
rating_matrix_2_demeaned = rating_matrix_2 - user_ratings_mean.reshape(-1, 1)
```

```python
print(np.mean(rating_matrix_2,axis=1))
```

```python
from scipy.sparse.linalg import svds
U, sigma, Vt = svds(rating_matrix_2_demeaned, k = 50)
```

```python
sigma = np.diag(sigma)
```

```python
all_user_predicted_ratings = np.dot(np.dot(U, sigma), Vt) + user_ratings_mean.reshape(-1, 1)
```

```python
print(all_user_predicted_ratings)
```

```python
preds_df = pd.DataFrame(all_user_predicted_ratings, columns = rating_matrix.columns)
```
